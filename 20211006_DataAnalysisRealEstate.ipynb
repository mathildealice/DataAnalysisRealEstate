```python
#cleaning up locality and postcode 
import pandas as pd
import numpy as np
import re

df = pd.read_csv('original_dataset.csv', low_memory = False, dtype = {'locality': str, 'postcode': str, 'property_subtype': str})
test_df = df


if 'loc_clean' not in test_df.columns:
    test_df.insert(3, "loc_clean", "clean value")
if 'loc_dirty' not in test_df.columns:
    test_df.insert(4, "loc_dirty", "dirty value")

#For all pc in locality, empty locality end copy value to clean column, if format is different, copy it to dirty column
pattern = re.compile("^\d{4}$")
row_number = -1
for row in test_df["locality"]: 
    row_number+=1
    if type(row) is str: 
        search_result = pattern.search(row)
        if search_result is None: 
            test_df.at[row_number, 'loc_dirty'] = row
        else: 
            if row == test_df["postcode"][row_number]: 
                test_df.loc[row_number, 'locality'] = np.nan
            test_df.at[row_number, 'loc_clean'] = row
test_df[test_df["loc_dirty"] == "dirty value"]
#takes a crazy amount of time
```
