```python
#Importing data
import pandas as pd
import numpy as np
import re

df = pd.read_csv('original_dataset.csv', low_memory = False, dtype = {'locality': str, 'postcode': str, 'property_subtype': str})
test_df = df

#Adding two column for data processing
if 'loc_clean' not in test_df.columns:
    index = test_df.columns.get_loc('locality')+1
    test_df.insert(index, "loc_clean", "clean value")
if 'loc_intermediate' not in test_df.columns: 
    index = test_df.columns.get_loc('loc_clean')+1
    test_df.insert(index, "loc_intermediate", "intermediate value")
if 'loc_dirty' not in test_df.columns:
    index = test_df.columns.get_loc('loc_intermediate')+1
    test_df.insert(index, "loc_dirty", "dirty_value")
    
test_df.head()
```

```python
#Catch post codes and set them in the intermediate column, set the rest in the dirty box
pattern = re.compile("^\d{4}$")
row_number = -1
for row in test_df["locality"]: 
    row_number+=1
    if type(row) is str: 
        search_result = pattern.search(row)
        if search_result is None: 
            test_df.at[row_number, 'loc_dirty'] = row
        else: 
            test_df.at[row_number, 'loc_intermediate'] = row
            test_df.loc[row_number, 'loc_clean'] = np.nan
            
test_df.head()
```

```python
row_number = -1
for row in test_df["locality"]:
    row_number+=1
    if row == test_df["postcode"][row_number]: 
        test_df.loc[row_number, 'locality'] = np.nan
        
test_df.head()
```

```python
import re
filter_pattern = re.compile("\d{4}")
pattern = re.compile("([a-zA-ZÀ-ÿ-,]+ ){0,}([a-zA-ZÀ-ÿ]+)$")
row_number = -1
for row in test_df["locality"]: 
    row_number+=1
    if type(row) is str: 
        filter_criteria = filter_pattern.search(row)
        if filter_criteria is not None: 
            search_result = pattern.search(row)
            if search_result is not None: 
                search_match = pattern.search(row).group()
                test_df.loc[row_number, 'loc_intermediate'] = search_match

#I think I should work with subsets of the data instaed of a filter pattern
#However this would mess up with my row_number

test_df.head()
```

```python
#Catch value before the pc 
import re

if 'address' not in test_df.columns:
    index = test_df.columns.get_loc('locality')
    test_df.insert(index, "address", "")
    
filter_pattern = re.compile("(, ){0,1}\d{4}")
pattern = re.compile("^([a-zA-ZÀ-ÿ -]+){1,}([0-9]+ ){0,}([a-zA-ZÀ-ÿ0-9]+){0,}")
row_number = -1
for row in test_df["locality"]: 
    row_number+=1
    if type(row) is str: 
        filter_criteria = filter_pattern.search(row)
        if filter_criteria is not None: 
            search_result = pattern.search(row)
            if search_result is not None: 
                search_match = pattern.search(row).group()
                test_df.loc[row_number, 'address'] = search_match
                
test_df.head()
```
